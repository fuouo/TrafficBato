### ROLLING 

```python
def seperateRollingWindows(TYPE, DATASET_DIVISION):
    WINDOWS = [2,3,4,8,12,24,32,48,96]
    for i in WINDOWS:
        WINDOWSIZE = i
        FILENAME = "eng_" + TYPE + "_" + ROAD + "_" + YEAR + "_" + DATASET_DIVISION

        raw_data = pd.read_csv("../../../../datasets/Engineered/Rolling/" + YEAR + "/" + DATASET_DIVISION + "/" + FILENAME + EXT, skipinitialspace=True)
        raw_data = raw_data.fillna(0)

        dataset = raw_data
        DONT_ERASE_WINDOW = WINDOWSIZE

        to_remove = ["Unnamed"]
        for s in WINDOWS: #2-9
            if s != DONT_ERASE_WINDOW:
                to_remove.append("=" + str(s))

        print(to_remove)

        cols_to_remove = []
        cols = dataset.columns
        for c in range(len(cols)):
            for r in to_remove:
                if len(cols[c].split("=")) > 1:
                    if str(r) == ("=" + cols[c].split("=")[1]):
                        print(str(r) + " in " + cols[c])
                        cols_to_remove.append(c)

        dataset = dataset.drop(dataset.columns[[cols_to_remove]], axis=1)

        ##################### EXPORTING TO CSV #####################
        if DATASET_DIVISION:
            FILENAME = "eng_win" + str(WINDOWSIZE) + "_" + TYPE + "_" + ROAD + "_" + YEAR + "_" + DATASET_DIVISION 

        print("exporting " + FILENAME)
        dataset.to_csv("../../../../datasets/Thesis Datasets/" + TYPE + "/" + FILENAME + EXT, encoding='utf-8', index=False)
        print("exported " + FILENAME + "\n\n")
        ############################################################

    print("--END--")
```

```python
seperateRollingWindows("mmda", "seasonWet_original")
```

### EXPANDING

```python
def seperateExpandingWindows(TYPE, DATASET_DIVISION):
    WINDOWS = [2,3,4,8,12,16,24,48,96]

    for i in WINDOWS:
        WINDOWSIZE = i

        if DATASET_DIVISION:
            FILENAME = "eng_" + TYPE + "_" + ROAD + "_" + YEAR + "_" + DATASET_DIVISION

        raw_data = pd.read_csv("../../../../datasets/Engineered/Expanding/" + YEAR + "/" + DATASET_DIVISION + "/" + TYPE + "/" + FILENAME + EXT, skipinitialspace=True)
        raw_data = raw_data.fillna(0)

        dataset = raw_data
        DONT_ERASE_WINDOW = WINDOWSIZE

        to_remove = []
        for s in WINDOWS: #2-9
            if s != DONT_ERASE_WINDOW:
                to_remove.append("Expanding Mean " + str(s))
                to_remove.append("Expanding Min " + str(s))            
                to_remove.append("Expanding Max " + str(s))
                to_remove.append("Expanding Sum " + str(s))

        print("Do not remove: " + str(DONT_ERASE_WINDOW))
        cols_to_remove = []
        cols = dataset.columns
        for c in range(1, len(cols)):
            t = cols[c].split(" ")
            if(t[0] == "Expanding"):
                for r in to_remove:
                    if(r == (t[0] + " " + t[1] + " " + t[2])):
                        cols_to_remove.append(c)
                        print(cols[c])

        dataset = dataset.drop(dataset.columns[[cols_to_remove]], axis=1)

        ##################### EXPORTING TO CSV #####################
        if DATASET_DIVISION:
            FILENAME = "eng_win" + str(WINDOWSIZE) + "_" + TYPE + "_" + ROAD + "_" + YEAR + "_" + DATASET_DIVISION 

        print("exporting " + FILENAME)
        dataset.to_csv("../../../../datasets/Engineered/Expanding/" + YEAR + "/" + DATASET_DIVISION + "/" + TYPE + "/" + 
                       ROAD + "/" + FILENAME + EXT, encoding='utf-8', index=False)
        print("exported " + FILENAME + "\n\n")
        ############################################################


    print("--END--")
```

### Rename Expanding Windows

```python
# Read the dataset
ROAD = "Vicente Cruz"
YEAR = "2015"
EXT = ".csv"
DATASET_DIVISION = "seasonWet"
TYPE = "mmda"
DIR = "../../../../datasets/Engineered/Expanding/" + YEAR + "/"
```

```python
WINDOWS = [2, 3, 4, 8, 12, 16, 24, 48, 96]

def renameExpanding(TYPE, DATASET_DIVISION):
    for w in WINDOWS:
        WEATHER_WINDOWS = w
        WEATHER_DIR = DIR + DATASET_DIVISION + "/" + TYPE + "/"
        WEATHER_FILENAME = "eng_" + TYPE + "_" + ROAD + "_" + YEAR + "_" + DATASET_DIVISION
        expanding_weather = pd.read_csv(WEATHER_DIR + WEATHER_FILENAME + EXT, skipinitialspace=True)
  
        if TYPE == "mmda":
            startIndex = 3
        elif TYPE == "wwo":
            startIndex = 13
        
        for c in range(startIndex, len(expanding_weather.columns)):
            col = expanding_weather.columns[c]
            colSplit = expanding_weather.columns[c].split(" ")
            print(colSplit)
            print(str(len(colSplit)))
            if len(colSplit) <= 1:
                continue
            
            expanding_weather = expanding_weather.rename(columns={col: colSplit[3] + "_E" + str.lower(colSplit[1]) + "=" + colSplit[2]})
        WEATHER_DIR = DIR + DATASET_DIVISION + "/" + TYPE + "/" + ROAD + "/"
        WEATHER_FILENAME = "eng_win" + str(WEATHER_WINDOWS) + "_" + TYPE + "_" + ROAD + "_" + YEAR + "_" + DATASET_DIVISION
        expanding_weather.to_csv(WEATHER_DIR + WEATHER_FILENAME + EXT, index=False, encoding='utf-8')
        print("Exporting to : " + WEATHER_DIR + WEATHER_FILENAME + EXT)
```

```python
renameExpanding("wwo", "seasonWet")
```

```python
renameExpanding("mmda", "SeasonWet")
```
