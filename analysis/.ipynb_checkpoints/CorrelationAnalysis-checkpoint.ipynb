```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from pandas.plotting import autocorrelation_plot
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
```

```python
def normalize(data):
    y = pd.to_numeric(data)
    y = np.array(y.reshape(-1, 1))
    
    scaler = MinMaxScaler()
    y = scaler.fit_transform(y)
    y = y.reshape(1, -1)[0]
    return y

def getAllHighCorr(dataset, limit):
    features = dataset.columns
    hc_features = []
    for w in range(0, len(features)):
        #print('Highly correlated with ' + w)
        col_name = features[w]
        col = dataset[col_name]
        temp = col[((col <= -limit) | (col >= limit))]
        hc_features.append(temp)
        #print(col[col>=0.5])
        #print('\n')
        
        
    for w in range(1, len(features)-1):
        hc_features[w] = hc_features[w].to_frame()
        print(hc_features[w])
        print('\n')
        
    return hc_features
```

## Importing Datasets

```python
ROADS_1 = ['A. Maceda', 'Anda Circle', 'Antipolo', 'Baclaran', 
         'Blumetritt', 'Buendia', 'Edsa Extension', 'Finance Road', 
         'Gov. Forbes', 'Lerma', 'Magsaysay Ave', 'P. Noval', 'Pablo Ocampo', 'Pedro Gil', 
         'Quezon Ave.', 'Quirino', 'Rajah Sulayman', 'Taft Ave.', 'U.N. Avenue',
         'Vicente Cruz']

ROAD = "Taft Ave."
YEAR = "2015"
EXT = ".csv"

DATASET_DIR = "../../../../datasets/"
```

#### Traffic Dataset

```python
TRAFFIC_WINDOWSIZE = 1
TRAFFIC_FILENAME = "mmda_" + ROAD + "_" + YEAR +  "_transformed"
traffic_raw_data = pd.read_csv(DATASET_DIR + "mmda/2015/transformed/" + TRAFFIC_FILENAME + EXT, skipinitialspace=True)
traffic_raw_data = traffic_raw_data.fillna(0)
traffic_dataset = traffic_raw_data

cols_to_remove =  [0, 1, 2]

traffic_dataset.drop(traffic_raw_data.columns[[cols_to_remove]], axis=1, inplace=True)
traffic_dataset.head()
```

#### Weather Dataset

```python
WEATHER_WINDOWSIZE = 2
WEATHER_FILENAME = "wwo_" + YEAR 
weather_dataset = pd.read_csv(DATASET_DIR +  "wwo/all-features/" + WEATHER_FILENAME + EXT, skipinitialspace=True)
weather_dataset = weather_dataset.fillna(0)

cols_to_remove = [0]
cols_to_remove += [2, 3, 13, 15, 17, 18, 21] #Remove Fahr and Miles features
weather_dataset = weather_dataset.drop(weather_dataset.columns[[cols_to_remove]], axis=1)
weather_dataset.head()
```

```python
merged_dataset = pd.merge(traffic_dataset, weather_dataset, left_index=True, right_index=True)
merged_dataset.head()
```

#### Correlation: Rainy Season - Dry Season - Whole Year (Traffic and Weather)

```python
#Analyzing correlation of month of June to October (Rainy Season)
weather_lag = 0
startMonth = 14496
endMonth = 29184
traffic = traffic_dataset.iloc[startMonth:endMonth,:]
traffic.statusS = traffic.statusS.rolling(96).mean()
weather = weather_dataset.iloc[startMonth-weather_lag:endMonth-weather_lag,:]
merged_dataset = pd.merge(traffic, weather, left_index=True, right_index=True)
mdc_rainy = merged_dataset.corr(method='spearman')
mdc = mdc_rainy
#CorrelationPlot
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(mdc, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(mdc),1)
ax.set_yticks(ticks)
ticks = np.arange(0,len(mdc.columns),1)
ax.set_xticks(ticks)
ax.set_yticklabels(mdc)
plt.title("Rainy Season")
#======================================================================#
#Analyzing correlation of month of November to May (Dry Season)
weather_lag = 0
traffic = [traffic_dataset.iloc[0:14496,:], traffic_dataset.iloc[29184:35040,:]]# January to May, November to December
traffic = pd.concat(traffic)
traffic.statusS = traffic.statusS.rolling(96).mean()
weather = weather_dataset.iloc[startMonth-weather_lag:endMonth-weather_lag,:]
merged_dataset = pd.merge(traffic, weather, left_index=True, right_index=True)
mdc_dry = merged_dataset.corr(method='spearman')
mdc = mdc_dry
#CorrelationPlot
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(mdc, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(mdc),1)
ax.set_yticks(ticks)
ticks = np.arange(0,len(mdc.columns),1)
ax.set_xticks(ticks)
ax.set_yticklabels(mdc)
plt.title("Dry Season")
#======================================================================#
# Correlating one whole year
merged_dataset = pd.merge(traffic_dataset.iloc[:,:], weather_dataset.iloc[:,:], left_index=True, right_index=True)
merged_dataset.statusS =merged_dataset.statusS.rolling(96).mean()
mdc_year = merged_dataset.corr(method='spearman')
mdc = mdc_year
#CorrelationPlot
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(mdc, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(mdc.columns),1)
ax.set_yticks(ticks)
ticks = np.arange(0,len(mdc.columns),1)
ax.set_xticks(ticks)
ax.set_yticklabels(mdc.columns)
plt.title("Whole Year")
plt.show()
```

```python
#Analyzing correlation of month of June to October (Rainy Season) with Lags
mdc_rainy_lags = []
weather_lag = range(0, 100)
startMonth = 14496
endMonth = 29184
for i in range(weather_lag):
    traffic = traffic_dataset.iloc[startMonth:endMonth,:]
    traffic.statusS = traffic.statusS.rolling(96).mean()
    weather = weather_dataset.iloc[startMonth-weather_lag:endMonth-weather_lag,:]
    merged_dataset = pd.merge(traffic, weather, left_index=True, right_index=True)
    mdc_rainy_lags.append(merged_dataset.corr(method='spearman'))
```
