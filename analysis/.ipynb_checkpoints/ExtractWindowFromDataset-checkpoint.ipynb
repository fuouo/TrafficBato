```python
import numpy as np
import pandas as pd
```

```python
# Read the dataset
ROAD = "Quirino"
YEAR = "2015"
EXT = ".csv"
TRANSFORMED = True

TYPE = "mmda" 
#TYPE = "wwo"

WINDOW_START = 2
WINDOW_END = 6
```

### Don't touch beyond this point :)

```python
# FOR TRAFFIC AND WEATHER

for i in range(WINDOW_START, WINDOW_END):
    WEATHER_WINDOWSIZE = i
    #engineered_noeng_wwo_Taft-Ave._2015
    if TRANSFORMED:
        WEATHER_FILENAME = "engineered_noeng_" + TYPE + "_" + ROAD + "_" + YEAR + "_transformed"
    else:
        WEATHER_FILENAME = "engineered_noeng_" + TYPE + "_" + ROAD + "_" + YEAR

    if TYPE == "wwo":
        WEATHER_FILENAME = "engineered_noeng_" + TYPE + "_" + ROAD + "_" + YEAR
        
    weather_raw_data = pd.read_csv("data/raw/" + WEATHER_FILENAME + EXT, skipinitialspace=True)
    weather_raw_data = weather_raw_data.fillna(0)

    weather_dataset = weather_raw_data
    DONT_ERASE_WINDOW = WEATHER_WINDOWSIZE

    to_remove = ["Unnamed"]
    for s in range(2, 10): #2-9
        if s != DONT_ERASE_WINDOW:
            to_remove.append("window = " + str(s))

    print(to_remove)

    cols_to_remove = []
    cols = weather_dataset.columns
    for c in range(len(cols)):
        for r in to_remove:
            if str(r) in cols[c]:
                print(str(c) + " = " + cols[c])
                cols_to_remove.append(c)

    weather_dataset = weather_dataset.drop(weather_dataset.columns[[cols_to_remove]], axis=1)

    ##################### EXPORTING TO CSV #####################
    if TRANSFORMED:
        FILENAME = "eng_win" + str(WEATHER_WINDOWSIZE) + "_" + TYPE + "_" + ROAD + "_" + YEAR + "_transformed"    
    else:
        FILENAME = "eng_win" + str(WEATHER_WINDOWSIZE) + "_" + TYPE + "_" + ROAD + "_" + YEAR
    
    print("exporting " + FILENAME)
    weather_dataset.to_csv("data/" + TYPE + "/" + FILENAME + EXT, encoding='utf-8', index=False)
    print("exported " + FILENAME + "\n\n")
    ############################################################
    
    
print("--END--")
```
